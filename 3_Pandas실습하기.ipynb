```python
# csv파일 D-F으로 불러오기

df_exam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/file/exam.csv')
df_exam
```

```python
# 기본 요약 통계  
df_exam.describe()
```

```python
# 상위 5개 미리보기
df_exam.head()
```

```python
# 하위 5개 미리보기
df_exam.tail()
```

```python
# D-F indexing[R에는 없엉] -> indexing  = [전체에서 부분을 뽑아내는 것]

# 1, LOC - column명 indexing
df_math = df_exam.loc[:,'math']
print(df_math)

df_eng = df_exam.loc[:,'english']
print(df_eng)

df_rs1 = df_exam.loc[1:4, ['math','english']]
print(df_rs1)

df_rs2 = df_exam.loc[4:10, ['math','english','science']]
print(df_rs2)

# 2. ILOC -  index 번호 indexing
df_rs3 = df_exam.iloc[1:5,2:4]
print(df_rs3)
```

```python
# 수학점수 60점 이상 D-F 출력하기
df_exam[df_exam.math>=60]
```

```python
# 수학점수 내림차순
df_exam.sort_values(by='math', ascending=False)
```

```python
# 영어점수 60 ~ 70 점 내림차순
df_exam[(df_exam.english >=60) & (df_exam.english < 70)].sort_values(by='english', ascending=False)
```

```python
# 파생변수
df_exam['total'] = df_exam.math + df_exam.english + df_exam.science
df_exam

df_exam['mean'] = df_exam.total / 3
df_exam
```
