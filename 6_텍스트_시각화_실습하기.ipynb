```python
# 형태소 분석
result1 = okt.pos('아버지가 방에 들어가십니다.')
print('result1 : ', result1)

result2 = okt.nouns('드론 만큼 매우 화난다.')
print('result2 : ', result2)
```

```python
# 텍스트 파일 불러오기
sample = open('/content/drive/MyDrive/Colab Notebooks/file/wordcloud_data.txt').read()
sample
# 명사만 따로 출력
sample_nouns = okt.nouns(sample)

# list인 sample을 String으로 변환
sample_nouns = " ".join(sample_nouns)

#단어구름 출력
wc = WordCloud(width=1200,
               height=1200,
               font_path='/content/drive/MyDrive/Colab Notebooks/file/HMFMPYUN.TTF',
               background_color = 'white').generate(sample_nouns)
pit.imshow(wc, interpolation='bilinear')
pit.show()
```
