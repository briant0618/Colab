```python
"""
날짜 : 2021/07/05
이름 : 김동현
내용 : Matplotlib 실습
"""

import matplotlib.pyplot as pit
import numpy as np
```

```python
# 기본 plotchart
pit.title('plot1')
pit.plot([1, 3, 4, 9, 10, 17])
pit.show
```

```python
# x축 설정 plotchart
pit.title('plot2')
pit.plot([10, 20, 30, 40, 50, 60],[1, 3, 4, 9, 10, 17])
pit.show
```

```python
# chart style
pit.title('plot3')
pit.plot([10, 20, 30, 40, 50, 60],[1, 3, 4, 9, 10, 17], 'bo-')
pit.grid(True)
pit.show
```

```python
# multi line chart
values = np.arange(0, 5, 0.2)


pit.title('plot4')
pit.plot(values, values, 'b--')
pit.plot(values, values ** 2, 'g--')
pit.plot(values, values ** 3, 'r--')
pit.grid(True)
pit.show
```

```python
# bar graph
pit.title('plot5')
pit.bar(['apple','banana','orange','grape','berry'],[5, 3, 7, 9, 8])
pit.xlabel('fruit')
pit.ylabel('count')
pit.show()
```

```python
# histogram
data1 = np.random.rand(1000) # 0 ~ 1 사이의 임의의 실수 1000개
data2 = np.random.randn(1000) # 0 ~ 1 사이의 표준정규분포를 따르는 임의의 실수 1000개



pit.title('histograph')
pit.hist(data1, edgecolor='black')
pit.show()
```

```python
pit.title('histograph')
pit.hist(data2, edgecolor='black')
pit.show()
```

```python
# scatter plot
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000) 

pit.title('scatter')
pit.scatter(x, y)
pit.show()
```
